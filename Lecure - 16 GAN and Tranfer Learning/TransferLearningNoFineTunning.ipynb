```python
import os
import random
import numpy as np
from keras.applications.resnet50 import ResNet50
from keras.preprocessing import image
import matplotlib.pyplot as plt
from keras.utils import  to_categorical
from keras.models import Model
from keras.layers import GlobalAveragePooling2D,Dense,Flatten,Input,Dropout
from keras.optimizers import Adam
```

```python
dirs = os.listdir('./data/')
```

```python
print(dirs)
```

```python
images = []
labels = []
```

```python
label_dict = {"cats":0,"dogs":1,"horses":2,"Humans":3}
```

```python
for ix in dirs:
    img_path = "./data/"+ix+"/"
    img_name = os.listdir(img_path)
    for i in img_name:
        img = image.load_img(img_path+i,target_size=(224,224))
        img_array = image.img_to_array(img)
        images.append(img_array)
        labels.append(label_dict[ix])
```

```python
print(len(images),len(labels))
```

```python
combined = list(zip(images,labels))
random.shuffle(combined)
images[:],labels[:] = zip(*combined)
```

```python
X_train = np.array(images)
Y_train = np.array(labels)
```

```python
y_train = to_categorical(Y_train)
```

```python
y_train
```

```python
res_model = ResNet50(include_top=False,weights = "imagenet",input_shape=(224,224,3))
```

```python
res_model.summary()
```

```python
av1 = GlobalAveragePooling2D()(res_model.output)
fc1 = Dense(256,activation='relu')(av1)
d1 = Dropout(0.25)(fc1)
out = Dense(4,activation='softmax')(d1)

t1_model = Model(inputs= res_model.input,outputs=out)
t1_model.summary()
```

```python
t1_model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
hist = t1_model.fit(X_train,y_train,epochs=5,shuffle=True,batch_size=64,validation_split=0.2)
```
